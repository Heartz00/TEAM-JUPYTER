DATA CLEANING OF THE HCAHPS SURVEY DATASETS

```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# DOWNLOAD THE DATASETS
ds_2016 =pd.read_csv('/content/drive/MyDrive/archive (2)/cms_hospital_patient_satisfaction_2016.csv')
ds_2017 = pd.read_csv('/content/drive/MyDrive/archive (2)/cms_hospital_patient_satisfaction_2017.csv')
ds_2018 = pd.read_csv('/content/drive/MyDrive/archive (2)/cms_hospital_patient_satisfaction_2018.csv')
ds_2019 = pd.read_csv('/content/drive/MyDrive/archive (2)/cms_hospital_patient_satisfaction_2019.csv')
ds_2020 = pd.read_csv('/content/drive/MyDrive/archive (2)/cms_hospital_patient_satisfaction_2020.csv')
```

```python
#check for missing values
ds_2016.isnull().sum()
ds_2017.isnull().sum()
ds_2018.isnull().sum()
ds_2019.isnull().sum()
ds_2020.isnull().sum()
```

Addressing the missing values

```python
# replacing the missing values with 'N', as the column is a yes or no feature
ds_2016['Meets criteria for promoting interoperability of EHRs'].fillna('N', inplace=True)
ds_2017['Meets criteria for promoting interoperability of EHRs'].fillna('N', inplace=True)
ds_2018['Meets criteria for promoting interoperability of EHRs'].fillna('N', inplace=True)
ds_2019['Meets criteria for promoting interoperability of EHRs'].fillna('N', inplace=True)
ds_2020['Meets criteria for promoting interoperability of EHRs'].fillna('N', inplace=True)
```

```python
#drop county Name column, due to its irrelevancies in the data
ds_2016.drop('County Name', axis=1, inplace=True)
ds_2017.drop('County Name', axis=1, inplace=True)
ds_2018.drop('County Name', axis=1, inplace=True)
ds_2019.drop('County Name', axis=1, inplace=True)
ds_2020.drop('County Name', axis=1, inplace=True)
```

```python
#drop all footnotes column, as it only add unwarranted noise to our data
drop_list = ['Patient Survey Star Rating Footnote','HCAHPS Answer Percent Footnote','Number of Completed Surveys Footnote',
            'Survey Response Rate Percent Footnote','Hospital overall rating footnote',
             'Mortality national comparison footnote','Safety of care national comparison footnote',
            'Readmission national comparison footnote','Patient experience national comparison footnote',
             'Effectiveness of care national comparison footnote', 'Timeliness of care national comparison footnote',
            'Efficient use of medical imaging national comparison footnote']

year_df = [ds_2016, ds_2017, ds_2018, ds_2019, ds_2020]
for year in year_df:
  year.drop(drop_list, axis=1, inplace=True)
```

Rechecking our data for missing values

```python
ds_2016.isnull().sum()
ds_2017.isnull().sum()
ds_2018.isnull().sum()
ds_2019.isnull().sum()
ds_2020.isnull().sum()
```

```python
#checking the size of the data
ds_2016.shape
ds_2017.shape
ds_2018.shape
ds_2019.shape
ds_2020.shape
```

Now our data is cleaned . We can save our new data for further analysis

```python
#save the cleaned data
ds_2016.to_csv('2016cleaned.csv')
ds_2017.to_csv('2017cleaned.csv')
ds_2018.to_csv('2018cleaned.csv')
ds_2019.to_csv('2019cleaned.csv')
ds_2020.to_csv('2020cleaned.csv')
```
